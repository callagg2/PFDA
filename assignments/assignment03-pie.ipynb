```python
# i will using pandas to manipulate my data, numpy for numerical operations, and matplotlib to create the pie chart
import pandas as pd

import numpy as np  

import matplotlib.pyplot as plt
```

```python
URL = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"
# this reads in the data from a csv file at a URL into a pandas dataframe
df = pd.read_csv(URL) 

print(f"{df.head()}\n")

# if there are any null values, put in blanks
cleandf= df.fillna(value=' ')

# add a column at the end that comprises the individual's email, 
# except we are splitting the email string either side of the "@"
# and taken the second part, that is position 1, whereas the first part is position 0.
cleandf['domain'] = cleandf['Email'].str.split('@').str[1]

# this just print out the first few rows of the dataframe so we can see the domain column is appearing
print(f"{cleandf['domain'].head()}")
```

```python
#unique, counts = np.unique(['domain'], return_counts=True)

# we can now put this into a pie plot
# The following method was taken from https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_and_donut_labels.html#sphx-glr-gallery-pie-and-polar-charts-pie-and-donut-labels-py

# we are telling Matplotlib to take the value counts of  the domain column
# assign labels to the unique labels, and finally to format the percentages to one decimal place
plt.pie(cleandf['domain'].value_counts(), labels=cleandf['domain'].unique(), autopct='%1.1f%%')

# this adds a title to the chart
plt.title("A Pie-Chart Showing the Breakdown of Email Domains")

# this add a legend to the pie-chart
plt.legend(labels=cleandf['domain'].unique(), title="Email Domains",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

#plt.savefig("pie-chart of email domains.png")

plt.show()
```
